```python
import pandas as pd
import json
import matplotlib.pyplot as plt

# csv paths
credits_path = "./data/tmdb_5000_credits.csv"
movies_path = "./data/tmdb_5000_movies.csv"

# create DataFrames
df_credits = pd.read_csv(credits_path)
df_movies = pd.read_csv(movies_path)
```

```python
df_credits.head()
```

```python
df_movies.head()
```

```python
df_credits.columns = ["id", "title", "cast", "crew"]

# merge both DataFrames
df_movies = df_movies.merge(df_credits, on="id")
df_movies.head()
```

```python
print(df_movies["genres"][0])
```

```python
from helper import extract_genre_names

# function on seires
df_movies['genre_names'] = df_movies['genres'].apply(extract_genre_names)

df_movies['genre_names'].head(3)
```

```python
plt.figure(figsize=(15, 5))
df_movies["original_language"].value_counts().plot(kind="bar", title="Original Language Value Counts", ylabel="Movies")
plt.show()
```

```python
# only English Movies

print(len(df_movies))
df_movies = df_movies[df_movies["original_language"] == "en"]
print(len(df_movies))
```

```python
# most important columns
df_movies = df_movies[["original_title", "genre_names", "overview", "release_date", "runtime", "vote_average", "vote_count", "popularity"]]
df_movies.head()
```

```python
# check NaN's and drop

print(df_movies.isna().sum())
df_movies = df_movies.dropna()
print("\n\n", df_movies.isna().sum())
```

```python
# check dtypes

df_movies.info()
```

```python
# turn runtime column from float64 to int

df_movies["runtime"] = df_movies["runtime"].astype(int)
```

```python
# turn release_date column from object to datetime

df_movies["release_date"] = pd.to_datetime(df_movies["release_date"])
```

```python
df_movies.head()
```

```python
# get Mean from vote_average Column

C = df_movies['vote_average'].mean()
C
```

```python
# get 0.9 quantile from vote_count

m = df_movies['vote_count'].quantile(0.9)
m
```

```python
# filter DataFrame

q_movies = df_movies.copy().loc[df_movies['vote_count'] >= m]
q_movies.shape
```

```python
from helper import weighted_rating

# Define a new feature 'score' and calculate its value with `weighted_rating()`
q_movies['score'] = q_movies.apply(weighted_rating(m=m, C=C), axis=1)
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies.head(10)
```

```python
from helper import create_summary

# create new column "summary"
df_movies['summary'] = df_movies.apply(create_summary, axis=1)

print(df_movies['summary'][0])
```

```python
# save preprocesses DataFrame as csv

df_movies.to_csv("tmdb_5000_prep.csv", index=False)
```
